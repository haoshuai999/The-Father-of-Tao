# Crawl water quality data now!
The following function is used to crawl data after July, 2004. The variable 'record_time' storages the date when the report was accomplished. The function is complex, because sometimes heavy rain, typhoon or other natural disasters stop the observation. The words such as "influeced by typhoon" take up several columns and the data structure is changed, so I have to change the code.

```python
def Crawler1(url):
    html = requests.get(url)
    html.encoding = 'utf-8'
    selector = etree.HTML(html.text)
    record_time = selector.xpath('normalize-space(//div[@class="wzxq_xia"]/div[@class="wzxq_biaoti2"]/span/p/text())').replace(' ', '').replace('科技标准司', '')
    content_field = selector.xpath('//tbody')
    for each in content_field:
        list = each.xpath('tr')
        temp_city = ''
        for tr in range(len(list)):
            beach = {}
            if(len(list[tr]) == 5):
                if (list[tr][0].xpath('string(.)').strip().replace(' ', '').replace(' ','') == '城市'):
                    continue
                temp_city = list[tr][0].xpath('string(.)').strip().replace(' ', '').replace(' ','')
                beach['city'] = temp_city
                beach['name'] = list[tr][1].xpath('string(.)').strip()
                beach['water_quality']= list[tr][2].xpath('string(.)').strip()
                beach['swim'] = list[tr][3].xpath('string(.)').strip().replace(' ', '').replace(' ','')
                beach['pollution'] = list[tr][4].xpath('string(.)').strip()

            elif(len(list[tr]) == 4):
                beach['city'] = temp_city
                beach['name'] = list[tr][0].xpath('string(.)').strip()
                temp_quality = beach['water_quality']= list[tr][1].xpath('string(.)').strip()
                beach['swim'] = list[tr][2].xpath('string(.)').strip().replace(' ', '').replace(' ','')
                beach['pollution'] = list[tr][3].xpath('string(.)').strip()
            elif(len(list[tr]) == 3):
                beach['city'] = list[tr][0].xpath('string(.)').strip()
                beach['name'] = list[tr][1].xpath('string(.)').strip()
                temp_quality = beach['water_quality']= beach['swim']= beach['pollution'] = list[tr][2].xpath('string(.)')
            elif(len(list[tr]) == 2):
                beach['city'] = temp_city
                beach['name'] = list[tr][0].xpath('string(.)').strip()
                beach['water_quality'] = beach['swim'] = temp_quality
                beach['pollution'] = list[tr][1].xpath('string(.)')
            elif (len(list[tr]) == 1):
                beach['city'] = temp_city
                beach['name'] = list[tr][0].xpath('string(.)').strip()
                beach['water_quality'] = beach['swim'] = beach['pollution'] = temp_quality
            elif (len(list[tr]) == 0):
                continue
            else:print(url)
                
            City.append(beach['city'])
            Name.append(beach['name'])
            Water_Quality.append(beach['water_quality'])
            Swim.append(beach['swim'])
            Pollution.append(beach['pollution'])
            Time.append(record_time)
```
