正则表达式符号与方法<br/>
findall  匹配所有符合查找规律的内容，返回的是一个包含所有结果的**列表**<br/>
search  匹配并提取第一个符合规律的内容，返回一个**正则表达式对象**<br/>
sub  替换符合规则的内容，返回**替换后的值**<br/>
- 导入模块包的方式：<br/>
    - import re <br/>
    - from re import * <br/>
    - from re import findall, search, sub, S<br/>

```python
pattern = 'xx(.*?)xx'
new_pattern = re.compile(pattern, re.S)
j = re.findall(new_pattern, string)
print j
```

#2. 正则表达式应用举例

**使用findall和search从大量文本中匹配目的内容**<br/>
**使用sub实现换页功能**<br/>
如果匹配多段内容，需要遵循“先抓大再抓小”的原则

如何实现翻页？<br/>

```python
old_url = "http://www.jikexueyuan.com/course/javascript/?pageNum=1"
for i in range(1, 11):
    new_link = re.sub('pageNum=\d+', 'pageNum=%d' %i, old_url)
    print new_link
```

```python
old = "http://sports.baidu.com/n?cmd=4&class=nba&pn=1"
for i in range(1, 21):
    new = re.sub('pn=(\d+)', 'page=%d'%i, old)
    print new
```

#3. 制作图片爬虫

目标网站：http://www.jikexueyuan.com/course/javascript/?pageNum=1<br/>
目标内容：所有课程的封面图片<br/>
实现原理：保存源代码至文件—>Python读取文件加载源码—>正则表达式提取图片网址—>下载图片

```python
import re, requests
old_url = "http://www.jikexueyuan.com/course/javascript/?pageNum=1"
f = requests.get(old_url)
print f.headers["set-cookie"] + '\n'
print str(f.status_code) + '\n'
print f.text
```

```python
import re
import requests
old_url = "http://www.jikexueyuan.com/course/javascript/?pageNum=1"

lst = []
for i  in range(1, 3):
    new_url  = re.sub('pageNum=(\d+)', 'pageNum=%d'%i, old_url)
    content = requests.get(new_url)
    div_1 = re.findall('<div class="lessonimg-box">(.*?)</div>', content.text, re.S)
    for item in div_1:
        pic_div = re.search('<img src="(.*?)" class="lessonimg"', item, re.S).group(1)
        lst.append(pic_div)
print lst

num = 1
for link in lst:
    print '正在下载……' + str(link)
    pic = requests.get(link)
    filepath = open('C:/Users/Luo Chen/Desktop/pics/' + str(num) + '.jpg', 'wb')            
    filepath.write(pic.content)
    filepath.close()
    num += 1
```
